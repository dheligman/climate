```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import dates
import seaborn as sns
```

```python
#connect to the plot tree species database
conn = sqlite3.connect('./oh_plot_tree_data/plot_tree_species2.db')
```

```python
#grabbing the tables for red maple and loblolly pine (the top 2 trees by population)
red_maple = pd.read_sql_query('SELECT * from "Red Maple"', conn)
loblolly_pine = pd.read_sql_query('SELECT * from "Loblolly Pine"', conn)
conn.close()
```

```python
#drop all the rows that have empty elements
red_maple = red_maple.dropna()
loblolly_pine = loblolly_pine.dropna()
```

```python
#creating columns for the date of measurements
red_maple['date'] = pd.to_datetime({'year':red_maple['measurement_year'], 
                                    'month':red_maple['measurement_month'],
                                   'day': np.ones(red_maple["measurement_year"].size)})
loblolly_pine['date'] = pd.to_datetime({'year':loblolly_pine['measurement_year'], 
                                    'month':loblolly_pine['measurement_month'],
                                   'day': np.ones(loblolly_pine["measurement_year"].size)})
```

```python
#creating dataframes for ohio
red_maple_oh = red_maple.loc[red_maple.plot_state_code == 39]
loblolly_pine_oh = loblolly_pine.loc[loblolly_pine.plot_state_code == 39]

#plotting the data for ohio to see what it looks like
X = red_maple_oh['date']
y = red_maple_oh['avg_height']
plt.figure(figsize = (10,10))
plt.scatter(X,y)
plt.show()
```

```python
#function for creating the dates for the x-axis for regression plots
@plt.FuncFormatter
def fake_dates(x, pos):
    """ Custom formater to turn floats into e.g., 2016-05-08"""
    return dates.num2date(x).strftime('%Y-%m-%d')
```

```python
#formatting the dates so that they can be plotted with a regression model
#and creating a new dataframe with a column that contains the reformatted dates
date_maple = dates.datestr2num(red_maple_oh.date.astype(str))
red_maple_oh_dtn = red_maple_oh.copy()
red_maple_oh_dtn['dtn'] = date_maple

date_loblolly = dates.datestr2num(loblolly_pine_oh.date.astype(str))
loblolly_pine_oh_dtn = loblolly_pine_oh.copy()
loblolly_pine_oh_dtn['dtn'] = date_loblolly
```

```python
#regression plot with 99% CI for the average height of red maples
fig, ax = plt.subplots(figsize = (10,10))

sns.regplot(data = red_maple_oh_dtn, x="dtn", y="avg_height",
           line_kws={"color": "red"},
            scatter_kws={"color": "green"},
           ci = 99)

plt.xlabel("Date", fontsize=16)
plt.ylabel("Average Height", fontsize=16)

#this converts the reformatted dates back to real dates for the graph
ax.xaxis.set_major_formatter(fake_dates)


plt.show()
```

```python
#plot for diameter for red maples
fig, ax = plt.subplots(figsize = (10,10))


sns.regplot(data = red_maple_oh_dtn, x="dtn", y="avg_diameter",
           line_kws={"color": "red"},
            scatter_kws={"color": "green"},
           ci = 99)

plt.xlabel("Date", fontsize=16)
plt.ylabel("Average Diameter", fontsize=16)

ax.xaxis.set_major_formatter(fake_dates)


plt.show()
```

```python
#plot for height of loblolly pines
#data is lacking for ohio :(
fig, ax = plt.subplots(figsize = (10,10))


sns.regplot(data = loblolly_pine_oh_dtn, x="dtn", y="avg_height",
           line_kws={"color": "red"},
            scatter_kws={"color": "green"},
           ci = 99)

plt.xlabel("Date", fontsize=16)
plt.ylabel("Average Height", fontsize=16)

ax.xaxis.set_major_formatter(fake_dates)


plt.show()
```
