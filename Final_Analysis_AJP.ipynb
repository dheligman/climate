```python
#data management imports
import numpy as np
import pandas as pd

#plotting imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

#misc python packages
from IPython.display import clear_output
import time
```

```python
files = ['RFPred.csv','SVMPred.csv','GNBPred.csv','MLPPred.csv','SoftEnsemblePred.csv']
models = ['Random\nForest', 'Support\nVector\nMachine', 'Gaussian\nNaive\nBayes', 
          'Multilayer\nLayer\nPerceptron', 'Soft\nEnsemble']
params = ['accuracy','recall','precision']
selection = pd.DataFrame({'Accuracy' : [],
                         'Recall' : [],
                         'Precision' : []})

for i, f in enumerate(files):
    temp = pd.read_csv(f)
    selection.loc[models[i]] = temp.iloc[-1][params].values
    
    
```

```python
selection
```

```python
selection.plot.bar(fontsize = 16,figsize = (16,7),legend = False, rot = 0)
plt.ylim((0,1.15))
plt.xlabel('',fontsize = 16)
plt.title('Model Comparison',fontsize = 16)
plt.legend(loc = 1,fontsize = 14)
plt.show()
```

```python
selection.to_csv('Model_Comparison.csv')
```
