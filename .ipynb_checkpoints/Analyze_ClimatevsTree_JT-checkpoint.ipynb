### What happens to tree volume and temperature in different states?

```python
states_array = county_tree_df.state_code.unique()
```

```python
state_code = pd.read_csv("County_State_Codes.csv")
state_name = state_code.loc[state_code.state_code==13]["state_name"].unique()[0]
```

```python
import corner
for states in states_array:
    county_1 = county_tree_df[(county_tree_df["state_code"]==states)]

    sweetgum_county1 = county_1.loc[county_1.species == 'sweetgum']
    sweetgum_county1["Biomass"] = sweetgum_county1["Biomass"]/sweetgum_county1["number_trees"]
    sweetgum_county1["Biomass"] = sweetgum_county1["Biomass"]/sweetgum_county1["number_trees"]
    sweetgum_county1 = sweetgum_county1.drop(["state_code", "county_code", "species", 'number_trees', "latitude",
                                             "longitude"], axis = 1)
    sweetgum_county1 = sweetgum_county1[(sweetgum_county1 != 0).all(1)]

    sweetgum_county1.head()
    samples = np.column_stack((sweetgum_county1["volume"], sweetgum_county1["mean_temp"]))
    figure = plt.figure(figsize=(4.9,6))
    figure = corner.corner(samples, labels=[r"volume_%s"%state_name, r"mean_temp_%s"%state_name],
                           quantiles=[0.16, 0.5, 0.84],
                           show_titles=True, title_kwargs={"fontsize": 12}, color='blue')
    state_name = state_code.loc[state_code.state_code==states]["state_name"].unique()[0]
    figure.suptitle("Species: sweetgum, state: %s"%state_name, y=1.03, fontsize=15) 

    figure.tight_layout()
    figure.savefig("/Users/neutrino/Downloads/Sweetgum_volume_vs_temp_%s.png"%state_name, dpi=150)
```
