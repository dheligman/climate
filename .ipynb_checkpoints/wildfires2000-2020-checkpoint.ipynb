```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
```

```python
fires = pd.read_csv('wildfires_00_20.csv')
fires['Number of Fires'] = fires['Number of Fires'].str.replace(',','').astype(float)
fires['Acres Burned'] = fires['Acres Burned'].str.replace(',','').astype(float)
fires = fires.astype(float)
fires['Months'] = np.linspace(0,fires.count()[0]-1,fires.count()[0])
fires.head()
```

```python
x = fires['Months']
y1, y2, y3 = fires['Number of Fires'], fires['Acres Burned'], fires['Acres Burned per Fire']
fig, axis1 = plt.subplots(figsize = (10,10))
axis1.plot(x, y1, '-r')
axis2 = axis1.twinx()
axis2.plot(x, y3, '-b')
axis1.set_xlabel('Month #', fontsize = 16)
plt.xlim(0,50)
axis1.set_ylabel('Number of Fires', fontsize = 16, color = 'r')
axis2.set_ylabel('Acres Burned per Fire', fontsize = 16, color = 'b')
plt.show()
```
