```python
def get_avg(df):
    
    measurement_year = df['measurement_year']
    species_code = hocking['species_code']
    avg = pd.DataFrame()
    years = measurement_year.sort_values().unique()
    species = species_code.dropna().sort_values().unique()
    avg['species_code'] = species
    
    for i in years:
        temp_avg_array = np.empty(len(species))
        for j in species:
            temp = df.loc[(df.species_code == j) & (df.measurement_year ==i)].mean().current_diameter
            temp_avg_array[np.where(species == j)] = temp
        avg[i] = temp_avg_array
    
    return avg
            
            
        
        
    
```
