```python
import requests
```

```python
#This is a script to grab a csv file if
#clicking on the file just leads to a 
#webpage csv instead of downloading it

csv_url = "https://www.ncdc.noaa.gov/societal-impacts/wildfires/month/0.csv"

req = requests.get(csv_url)

url_content = req.content

csv_file = open('wildfires_00_20.csv', 'wb')

csv_file.write(url_content)

csv_file.close()
```
